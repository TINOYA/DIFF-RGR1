```python
h=0.2
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams.update({
    'axes.facecolor': '1',
    'figure.facecolor': '0.8'
})


x = []
y = []
i = 0
while (i<=1):
  x.append(i)
  y.append(i**2 + 1)
  i+=0.000001

i = 0
x_a = []
y_a = []
while (i<=1):
  x_a.append(i)
  y_a.append(i**2 + 1)
  plt.scatter(i, i**2 + 1, color = 'b')
  plt.text(i, i**2 + 1 - 0.05, "("+str(round(i, 1))+"; "+str(round(i**2+1, 4))+')')
  i+=h


plt.plot(x, y)
plt.xlabel('х')
plt.ylabel('y')
plt.xlim([-0.1, 1.5])
plt.ylim([0.8, 2.1])

plt.show()
```

```python
def f(x, y):
  return (2*x*y)/(x**2 + 1)

xi = 0
yi = 1
x_ei = [xi]
y_ei = [yi]
while (xi<1):
  plt.text(xi, yi-0.05, "("+str(round(xi, 1))+"; "+str(round(yi, 4))+')')
  yi = yi + h*f(xi, yi)
  xi += h
  x_ei.append(xi)
  y_ei.append(yi)
plt.text(xi, yi-0.05, "("+str(round(xi, 1))+"; "+str(round(yi, 4))+')')

plt.plot(x_ei, y_ei, marker = 'o')
plt.xlabel('х')
plt.ylabel('y')
plt.xlim([-0.1, 1.5])
plt.ylim([0.8, 2.1])
```

```python
def f(x, y):
  return (2*x*y)/(x**2 + 1)

xi = 0
yi = 1
x_m_ei = [xi]
y_m_ei = [yi]
while (xi<1):
  plt.text(xi, yi-0.05, "("+str(round(xi, 1))+"; "+str(round(yi, 4))+')')
  y_half_i = yi + (h/2)*f(xi, yi)
  x_half_i = xi + h/2
  delta_y = h * f(x_half_i, y_half_i)
  xi += h
  yi += delta_y
  x_m_ei.append(xi)
  y_m_ei.append(yi)
plt.text(xi, yi-0.05, "("+str(round(xi, 1))+"; "+str(round(yi, 4))+')')

plt.plot(x_m_ei, y_m_ei, marker = 'o')
plt.xlabel('х')
plt.ylabel('y')
plt.xlim([-0.1, 1.5])
plt.ylim([0.8, 2.1])
```

```python
plt.plot(x_a, y_a, marker = 'o')
plt.plot(x_ei, y_ei, marker = 'o')
plt.plot(x_m_ei, y_m_ei, marker = 'o')
plt.xlabel('х')
plt.ylabel('y')
plt.xlim([-0.1, 1.5])
plt.ylim([0.8, 2.1])
plt.legend(["Аналитический", "Эйлера", "Мод. Эйлера"])
```
