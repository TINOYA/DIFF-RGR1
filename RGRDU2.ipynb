```python
h=0.1
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import math
mpl.rcParams.update({
    'axes.facecolor': '1',
    'figure.facecolor': '0.8'
})

def f_y(x):
  return 4*math.cos(x)+math.sin(x)+x**2-3

x = []
y = []
i = 0
while (i-0.3<=0.0000001):
  x.append(i)
  y.append(f_y(i))
  i+=0.000001

i = 0
x_a = []
y_a = []
while (i-0.3<=0.0000001):
  print(i)
  x_a.append(i)
  y_a.append(f_y(i))
  plt.scatter(i, f_y(i), color = 'b')
  plt.text(i, f_y(i) - 0.05, "("+str(round(i, 1))+"; "+str(round(f_y(i), 6))+')')
  i+=h


plt.plot(x, y)
plt.xlabel('х')
plt.ylabel('y')
plt.xlim([-0.1, 1])
plt.ylim([0.8, 1.5])

plt.show()
```

```python
def f1(x, y, z):
  return z

def f2 (x, y, z):
  return -y + x**2 - 1

xi = 0
yi = 1
zi = 1
x_r = [xi]
y_r = [yi]
while (xi<=0.3):
  plt.text(xi, yi-0.05, "("+str(round(xi, 1))+"; "+str(round(yi, 6))+')')
  k1 = h*f1(xi, yi, zi)
  l1 = h*f2(xi, yi, zi)

  k2 = h*f1(xi + h/2, yi + k1/2, zi + l1/2)
  l2 = h*f2(xi + h/2, yi + k1/2, zi + l1/2)

  k3 = h*f1(xi + h/2, yi + k2/2, zi + l2/2)
  l3 = h*f2(xi + h/2, yi + k2/2, zi + l2/2)

  k4 = h*f1(xi + h, yi + k3, zi + l3)
  l4 = h*f2(xi + h, yi + k3, zi + l3)


  delta_y = (1/6)*(k1 + 2*k2 + 2*k3 + k4)
  delta_z = (1/6)*(l1 + 2*l2 + 2*l3 + l4)

  xi += h
  yi += delta_y
  zi += delta_z
  print(round(k1, 6), round(l1, 6), round(k2, 6), round(l2, 6), round(k3, 6), round(l3, 6), round(k4, 6), round(l4, 6), delta_y, delta_z, yi, zi)

  x_r.append(xi)
  y_r.append(yi)
plt.text(xi, yi-0.05, "("+str(round(xi, 1))+"; "+str(round(yi, 6))+')')

plt.plot(x_r, y_r, marker = 'o')
plt.xlabel('х')
plt.ylabel('y')
plt.xlim([-0.1, 1])
plt.ylim([0.8, 1.5])
```
